```python
#实现池化层
import os 
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
import torch
```

```python
def my_pooling(input_tensor,pooling_size,mode='max'):
    """参数的含义如下："""
    """input_tensor:输入的张量
    pooling_size:池化窗口的大小
    mode:池化的模式，有max和avg两种
    """
    p_h,p_w=pooling_size
    Y=torch.zeros(input_tensor.shape[0],input_tensor.shape[1],input_tensor.shape[2]//p_h,input_tensor.shape[3]//p_w)
    for i in range(input_tensor.shape[0]):
        for j in range(input_tensor.shape[1]):
            if mode=='max':
                Y[i,j]=input_tensor[i,j].view(input_tensor.shape[2]//p_h,p_h,input_tensor.shape[3]//p_w,p_w).max(dim=1)[0].max(dim=2)[0]
            elif mode=='avg':
                Y[i,j]=input_tensor[i,j].view(input_tensor.shape[2]//p_h,p_h,input_tensor.shape[3]//p_w,p_w).mean(dim=1).mean(dim=2)
    return Y


input_tensor=torch.rand(1,3,32,32)
output_tensor=my_pooling(input_tensor,(2,2),'avg')
print(output_tensor)
```
